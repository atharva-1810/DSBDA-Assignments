```python
import pandas as pd
import numpy as np
#importing pandas and numpy
```

```python
name=['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location',
'wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system',
'bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
df= pd.read_csv("/content/auto.csv",names=name)
#importing data and adding related headers
```

```python
df["normalized-losses"]=pd.to_numeric(df["normalized-losses"])
df["stroke"]=pd.to_numeric(df["stroke"])
df["bore"]=pd.to_numeric(df["bore"])
df["horsepower"]=pd.to_numeric(df["horsepower"])
df["peak-rpm"]=pd.to_numeric(df["peak-rpm"])
df.dtypes
#change datatype to numeric
```

```python
mean_value = df["stroke"].mean()
df["stroke"].fillna(mean_value,inplace=True)
df.tail(10)
#replacing nan by mean
```

```python
mean_value = df["normalized-losses"].mean()
df["normalized-losses"].fillna(mean_value,inplace=True)
df.head(10)
#replacing nan by mean
```

```python
mean_value = df["bore"].mean()
df["bore"].fillna(mean_value,inplace=True)
df.head(10)
#replacing nan by mean
```

```python
mean_value = df["horsepower"].mean()
df["horsepower"].fillna(mean_value,inplace=True)
df.head(10)
#replacing nan by mean
```

```python
mean_value = df["peak-rpm"].mean()
df["peak-rpm"].fillna(mean_value,inplace=True)
df.head(10)
#replacing nan by mean
```

```python
df.isnull().sum()
```

```python
df['num-of-doors']=df['num-of-doors'].replace(np.nan,'four')
df.isnull().sum()
#replace by frequency
```

```python
df.dropna(subset=['price'],inplace=True)
#drop whole row - price
```

```python
len_m=df['length'].max()
wid_m=df['width'].max()
hgt_m=df['height'].max()
print(len_m,wid_m,hgt_m)
```

```python
df['length']=df['length']/len_m
df['width']=df['width']/wid_m
df['height']=df['height']/hgt_m
#normalization--scaling
```

```python
df['city-l/100']=235/df['city-mpg']
df['highway-l/100']=235/df['highway-mpg']
#standardization
```
