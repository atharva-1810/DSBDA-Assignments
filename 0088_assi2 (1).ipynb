```python
import pandas as pd
import numpy as np
df= pd.read_csv("/content/EduData.csv")
```

```python
df.replace('?',np.nan ,inplace=True)
df.isnull().sum()
```

```python
df['Topic'].fillna(df['Topic'].mode()[0],inplace=True)
#replacing with most common topic (mean) i.e. IT
```

```python
df["raisedhands"]=pd.to_numeric(df["raisedhands"])
df['raisedhands'].fillna(df['raisedhands'].mean(),inplace=True)
#replacing with average
```

```python
df = df.dropna(subset=['Relation',"GradeID"])
df.isnull().sum()
#Removing rows from relation,gradeID where value is nan
```

```python
df
```

```python
#outliers
freq_semester=df["Semester"].value_counts()
print(freq_semester)
```

```python
df["Semester"].replace(freq_semester.index[-1],freq_semester.index[0],inplace=True)
freq_semester=df["Semester"].value_counts()
print(freq_semester)
```

```python
df = df.drop('ParentAnsweringSurvey',axis=1)
#drop column ParentAnsweringSurvey
```

```python
#normalization of raisedhands and VisITedResources
df['raisedhands']=df['raisedhands']/df['raisedhands'].max()
df['VisITedResources']=df['VisITedResources']/df['VisITedResources'].max()
```

```python
#transforming gradeID
freq_GradeID = df["GradeID"].value_counts()
print(freq_GradeID.index.sort_values())
```

```python
Dict1 = {'G-02':1, 'G-04':2, 'G-05':3, 'G-06':4, 'G-07':5, 'G-08':6, 'G-09':7, 'G-10':8, 'G-11':9,
       'G-12':10}
```

```python
df["GradeID"] = df["GradeID"].map(Dict1)
df.head(10)
#Convert gradeID to '1-10'
```

```python
freq_abs_stu = df["StudentAbsenceDays"].value_counts()
print(freq_abs_stu.index.sort_values())
```

```python
#transforming StudentAbsenceDays
value2 = df["StudentAbsenceDays"].unique().tolist()
i=1
value2.sort()
for data in value2:
    df["StudentAbsenceDays"].replace(data,i,inplace=True)
    i -= 1
```

```python
df.head(10)
```

```python
#binning-equal frequency
num_bin=7
bin_label = ["extremely-low","very-low","low","average","high","very-high","extremely-high"]
df["Discussion-Binning"]=pd.qcut(df["Discussion"],q=num_bin,labels=bin_label)
df.head(10)
```

```python
#binning-equal-width
bin_label_equal_width=["O","A+","A","B+","B","C+","C"]
df["Discussion-equal-width-binning"]=pd.cut(df["Discussion"],bins=num_bin,labels=bin_label_equal_width)
df.head(10)
```

```python
#binning-custom
custom_edges=[22,32,44,54,63,71,88,94]
bin_custom_labels=["O","A","B","C","D","E","F"]
df["Discussion-custom-binning"]=pd.cut(df["Discussion"],bins=custom_edges,labels=bin_custom_labels)
df.head(10)
```
