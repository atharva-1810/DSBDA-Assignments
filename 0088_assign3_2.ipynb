```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("D:\ASSIGNMENTS\\3rd YEAR 2nd SEM\\DSBDA\\anti dataset - anti dataset.csv")
```

```python
df
```

```python
df["V1-Equal-width"] = pd.cut(df["V1"],5,labels=["very-low","low","medium","high","very-high"])
df
```

```python
df["V1-equal-freq"] = pd.qcut(df["V1"],5,labels=["A","B","C","D","E"])
df
```

```python
bins = [0,50,100]
labels_1=["low","high"]
df["equal-div"] = pd.cut(df["V1"],bins,labels=labels_1)
df
```

```python
low_df = pd.DataFrame(df.loc[df["equal-div"]=="low"])
low_df
```

```python
high_df = pd.DataFrame(df.loc[df["equal-div"]=="high"])
high_df
```

```python
#normalization
low_df["ANTI_A"] = (low_df["ANTI_A"] - low_df["ANTI_A"].min())/(low_df["ANTI_A"].max() - low_df["ANTI_A"].min())
low_df["ANTI_B"] = (low_df["ANTI_B"] - low_df["ANTI_B"].min())/(low_df["ANTI_B"].max() - low_df["ANTI_B"].min())
low_df
```

```python
plt.scatter(low_df["ANTI_A"],low_df["ANTI_B"])
plt.xlabel("ANTI_A")
plt.ylabel("ANTI_B")
plt.show()
```

```python
plt.scatter(high_df["ANTI_A"],high_df["ANTI_B"])
plt.xlabel("ANTI_A")
plt.ylabel("ANTI_B")
plt.show()
```
