```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv("D:\\DSBDA\\Assi7\\spam.csv",encoding="ISO-8859-1")
```

```python
dataframe = dataframe.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'])
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
corpus=[]
for i in range (0,5572):
    sms = re.sub("[^a-zA-Z]", " ", dataframe['v2'][i])
    sms = sms.lower()
    sms = sms.split()
    ps = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    all_stopwords = stopwords.words('english')
    sms = [ps.stem(word) for word in sms if not word in set(all_stopwords)]
    sms = [lemmatizer.lemmatize(word) for word in sms if not word in set(all_stopwords)]
    sms = " ".join(sms)
    corpus.append(sms)
```

```python
from nltk import pos_tag
from nltk import word_tokenize, sent_tokenize
tokens = word_tokenize(corpus[0])
print(pos_tag(tokens))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
X = tfidf_vectorizer.fit(corpus)
y = dataframe.iloc[:, -1].values
```
